# 실습 1.
### 2022년 어린이대공원(세종대)역 월별 승하차 인원 현황

https://www.data.go.kr/data/15044249/fileData.do

#### 1. 라이브러리 import

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

#### 2. csv 파일 불러오기

```python
df = pd.read_csv('서울교통공사_월별 승하차인원_20220531.csv', encoding='CP949')
df
```

#### 3. 어린이대공원(세종대)역 데이터 추출

```python
sejong = df.loc[df['역명'] == '어린이대공원'].iloc[:,4:]
sejong = list(map(int, sejong.iloc[0]))
sejong
```

```python
# 월 추출
month = df.columns[4:].tolist()

for i in range(len(month)):
    month[i] = month[i][5:]
    
month
```

#### 4. 그래프 그리기

```python
plt.rcParams["font.family"] = "Malgun Gothic"
plt.rc('font', size=20)

fig, ax = plt.subplots(figsize=(20,15))

label = np.arange(5)

bars = ax.bar(label, sejong, tick_label=month, align='center', color='gray')
ax.set_ylabel('승하차수 (명)')
ax.set_title('2022년 어린이대공원(세종대)역 월별 승하차 인원 현황', fontweight='bold', fontsize=30)

# plt.bar(label, sejong, tick_label=month, align='center', color='gray')
# plt.title('2022년 어린이대공원(세종대)역 월별 승하차 인원 현황')
# plt.ylabel('승하차수 (명)')

# 지수 표현 형식 없애기
current_values = plt.gca().get_yticks()
plt.gca().set_yticklabels(['{:.0f}'.format(x) for x in current_values])

for i in range(len(label)):
    height = sejong[i]
    if label[i] > 1:
        plt.text(label[i], height+1.0, '%d' %(height), ha='center', va='bottom', size=15, color='r')
    else:
        plt.text(label[i], height+1.0, '%d' %(height), ha='center', va='bottom', size=15)

bars[2].set_color("#FEE3EC")
bars[3].set_color("#F9C5D5")
bars[4].set_color("#F9C5D5")

plt.show()
```

# 실습 2.
### 2011년 ~ 2021년 운전면허소지자 성별 현황
https://www.data.go.kr/data/15048417/fileData.do

#### 1. 라이브러리 import

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

#### 2. csv 파일 불러오기

```python
df = pd.read_csv('경찰청_운전면허소지자 성별 현황_20211231.csv', encoding='CP949')
df
```

#### 3. 2011~2021년 운전면허 소지자 데이터 추출

```python
# 연도 추출
year = np.array(df['연도'].iloc[26:])
year
```

```python
# 남성 운전면허소지자 추출
male = np.array(df['남'].iloc[26:])
male
```

```python
# 여성 운전면허소지자 추출
female = np.array(df['여'].iloc[26:])
female
```

#### 4. 그래프 그리기

```python
plt.rcParams["font.family"] = "Malgun Gothic"
plt.rc('font', size=20)

y = np.arange(male.size)

fig, axes = plt.subplots(ncols=2, sharey=True, figsize=(25, 10))
axes[0].barh(y, male, align='center', color='b', zorder=10)
axes[0].set(title="남성 운전면허 소지자")

axes[1].barh(y, female, align='center', color='r', zorder=10)
axes[1].set(title="여성 운전면허 소지자")

axes[0].invert_xaxis()
axes[0].invert_yaxis()
axes[0].set(yticks=y, yticklabels=year)
axes[0].yaxis.tick_right()
axes[0].ticklabel_format(axis='x',useOffset=False, style='plain')
axes[0].set_xlabel('운전면허소지자(명)', fontsize=25)

axes[1].ticklabel_format(axis='x',useOffset=False, style='plain')
axes[1].set_xlabel('운전면허소지자(명)', fontsize=25)

for ax in axes.flat:
    ax.margins(0.03)
    ax.grid(True)

fig.tight_layout()
fig.subplots_adjust(wspace=0.09)

plt.show()
```

```python
# 혈액형
blood_group = df.iloc[1, 13:].tolist()
print(blood_group)

blood_group_size = list(map(int, df.iloc[2, 13:]))
print(blood_group_size)
```

# 실습 4.
### 서울시 자치구별 CCTV 설치 현황
https://data.seoul.go.kr/dataList/OA-2734/F/1/datasetView.do

#### 1. 라이브러리 import

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
!pip install folium
import folium
```

#### 2. csv 파일 불러오기

```python
df = pd.read_csv('서울시CCTV설치운영현황(자치구)_년도별_211231기준.csv', encoding='CP949')
df
```

#### 3. 서울시 자치구별 CCTV 설치 현황 데이터 추출

```python
# 서울 지도에서 행정 구역별 표시
import requests
import json

# 서울 행정구역 json raw파일(githubcontent)
r = requests.get('https://raw.githubusercontent.com/southkorea/seoul-maps/master/kostat/2013/json/seoul_municipalities_geo_simple.json')
c = r.content
seoul_geo = json.loads(c)
```

```python
# 서울시 자치구별 총 CCTV 데이터 추출
seoul = df.iloc[1:, 0:2]

# 중 구 -> 중구로 변경
seoul.loc[2, '구분'] = '중구'
seoul
```

```python
seoul.dtypes
```

```python
seoul['총계'] = seoul['총계'].str.replace(',', '').astype(float).astype('int64') # dtype: object -> int 변환
seoul.dtypes
```

#### 4. 그래프 그리기

```python
bins = list([0, 1000, 2000, 3000, 4000, 5000, 6000, 7000])

m = folium.Map(
    location=[37.559819, 126.963895],
    zoom_start=11, 
    tiles='cartodbpositron'
)

folium.GeoJson(
    seoul_geo,
    name='지역구'
).add_to(m)

m.choropleth(geo_data=seoul_geo,
             data=seoul,
             columns=['구분', '총계'],
             fill_color='YlGnBu', # 색상 변경도 가능하다
             fill_opacity=0.5,
             line_opacity=0.2,
             key_on='properties.name',
             legend_name="자치구별 CCTV 수", 
             bins=bins
            )
m
```

```python
m.save('seoul_cctv_map.html')
```

# 실습 5.
### 2021년 서울시 12개 자치구별 월별 교통사고 사고건수 현황
https://www.data.go.kr/data/15070295/fileData.do

#### 1. 라이브러리 import

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

#### 2. csv 파일 불러오기

```python
df = pd.read_csv('도로교통공단_시군구별 월별 교통사고 통계_20211231.csv', encoding='CP949')
df
```

#### 3. 2021년 서울시 12개 자치구별 월별 교통사고 사고건수 데이터 추출

```python
month = ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December']
month
```

```python
# 강남구
gangnam = df.loc[df['시군구'] == '강남구','사고건수']
gangnam.index = month
gangnam = gangnam.to_frame(name='강남구')
gangnam
```

```python
# 강동구
gangdong = df.loc[df['시군구'] == '강동구','사고건수']
gangdong.index = month
gangdong = gangdong.to_frame(name='강동구')
gangdong
```

```python
# 강서구
gangseo = df.loc[(df['시도'] == '서울') & (df['시군구'] == '강서구'),'사고건수']
gangseo.index = month
gangseo = gangseo.to_frame(name='강서구')
gangseo
```

```python
# 광진구
gwangjin = df.loc[df['시군구'] == '광진구','사고건수']
gwangjin.index = month
gwangjin = gwangjin.to_frame(name='광진구')
gwangjin
```

```python
# 노원구
nowon = df.loc[df['시군구'] == '노원구','사고건수']
nowon.index = month
nowon = nowon.to_frame(name='노원구')
nowon
```

```python
# 동대문구
dongdaemun = df.loc[df['시군구'] == '동대문구','사고건수']
dongdaemun.index = month
dongdaemun = dongdaemun.to_frame(name='동대문구')
dongdaemun
```

```python
# 동작구
dongjag = df.loc[df['시군구'] == '동작구','사고건수']
dongjag.index = month
dongjag = dongjag.to_frame(name='동작구')
dongjag
```

```python
# 마포구
mapo =df.loc[df['시군구'] == '마포구','사고건수']
mapo.index = month
mapo = mapo.to_frame(name='마포구')
mapo
```

```python
# 서초구
seocho = df.loc[df['시군구'] == '서초구','사고건수']
seocho.index = month
seocho = seocho.to_frame(name='서초구')
seocho
```

```python
# 송파구
songpa = df.loc[df['시군구'] == '송파구','사고건수']
songpa.index = month
songpa = songpa.to_frame(name='송파구')
songpa
```

```python
# 영등포구
yeongdeungpo = df.loc[df['시군구'] == '영등포구','사고건수']
yeongdeungpo.index = month
yeongdeungpo = yeongdeungpo.to_frame(name='영등포구')
yeongdeungpo
```

```python
# 중구
jung = df.loc[(df['시도'] == '서울') & (df['시군구'] == '중구'),'사고건수']
jung.index = month
jung = jung.to_frame(name='중구')
jung
```

```python
# 서울시 자치구별 월별 교통사고 사고건수 dataframe 만들기
tf = pd.DataFrame() # 빈 데이터프레임

tf = pd.concat([gangnam, gangdong, gangseo, gwangjin, nowon, dongdaemun, dongjag, mapo, seocho, songpa, yeongdeungpo, jung],axis=1)
tf
```

```python
# 월별 교통사고 평균 사고 건수
tf.mean(axis=1)
```

```python
# 자치구별 교통사고 평균 사고 건수
tf.mean(axis=0)
```

#### 4. 그래프 그리기

```python
plt.rcParams["font.family"] = "Malgun Gothic"
plt.rc('font', size=12)

fig = plt.figure(figsize=(10, 10))

# heatmap by seaborn
cmap = sns.cm.rocket_r # reverse
ax = sns.heatmap(tf, annot=True, fmt='d', cmap=cmap)
plt.title('2021년 서울시 12개 자치구별 월별 교통사고 사고건수', fontsize=20, fontweight='bold')

plt.show()
```

# 실습 6.
### 2020년 국민건강보험 가입자의 허리둘레별 총 콜레스테롤 비교
https://www.data.go.kr/data/15007122/fileData.do

#### 1. 라이브러리 import

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

#### 2. csv 파일 불러오기

```python
df = pd.read_csv('국민건강보험공단_건강검진정보_20211229.CSV', encoding='CP949')
df
```

#### 3. 2020년 국민건강보험 가입자들의 허리둘레별 콜레스트롤 수치 데이터 추출

```python
# 총 콜레스테롤 수치 결측치 제외
df = df.dropna(subset=['총 콜레스테롤'])
df
```

```python
# 2020년 국민건강보험 가입자 100명 무작위로 추출
df = df.sample(n=100)
df
```

```python
# 성별코드, 허리둘레, 총 콜레스테롤 추출
new_df = df[['성별코드', '허리둘레', '총 콜레스테롤']]
new_df
```

```python
male = df.loc[df['성별코드'] == 1]
male
```

```python
female = df.loc[df['성별코드'] == 2]
female
```

#### 4. 그래프 그리기

```python
plt.rcParams["font.family"] = "Malgun Gothic"
plt.rc('font', size=15) # 기본 폰트 크기

fig = plt.figure(figsize=(25,20))

area = ()

plt.scatter(male['허리둘레'], male['총 콜레스테롤'], color='b', alpha=0.4, label='남성', s =10*male['허리둘레'])
plt.scatter(female['허리둘레'], female['총 콜레스테롤'], color='r', alpha=0.2, label='여성', s = 10*female['허리둘레'], marker='s')

plt.xlabel('허리둘레(cm)', fontsize=20)
plt.ylabel('총 콜레스트롤(㎎/㎗)', fontsize=20)
plt.title("2020년 국민건강보험 가입자 100명의 허리둘레별 총 콜레스테롤 비교", size = 30)

plt.axhline(y=240, color='r', linewidth=2)
plt.text(58, 243, '높음', color='r', fontsize='18')
plt.axhline(y=200, color='g', linewidth=3)
plt.text(58, 203, '적정', color='g', fontsize='18')

plt.legend(fontsize=18)
plt.show()
```
