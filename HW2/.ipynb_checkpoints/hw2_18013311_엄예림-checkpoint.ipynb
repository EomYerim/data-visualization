# 실습 1.
### 2016년 ~ 2021년 서울시 SNS 유형별 이용 현황

#### 1. 라이브러리 import

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

#### 2. csv 파일 불러오기

```python
df = pd.read_csv('SNS+이용률_20221002040825.csv')
df
```

#### 3. 2016년 ~ 2021년 유형별 SNS 이용률 추출

```python
# 페이스북 - 2016~2021년 데이터 추출
facebook = df.loc[df['구분별(2)'] == '페이스북'].iloc[:, 6:]
facebook = list(map(float, facebook.iloc[0])) # str->float 및 dataframe -> list
facebook
```

```python
# 카카오스토리 - 2016~2021년 데이터 추출
kakaostory = df.loc[df['구분별(2)'].str.contains('카카오스토리')].iloc[:, 6:]
kakaostory = list(map(float, kakaostory.iloc[0]))
kakaostory
```

```python
# 인스타그램 - 2016~2021년 데이터 추출
instagram = df.loc[df['구분별(2)'].str.contains('인스타그램')].iloc[:, 6:]
instagram = list(map(float, instagram.iloc[0]))
instagram
```

```python
# 네이버밴드 - 2016~2021년 데이터 추출
band = df.loc[df['구분별(2)'].str.contains('네이버 밴드')].iloc[:, 6:]
band = list(map(float, band.iloc[0]))
band
```

#### 4. 그래프 그리기

```python
plt.rcParams["font.family"] = "Malgun Gothic"
plt.rc('font', size=20)

fig = plt.figure(figsize=(12, 15))

x_values = df.columns[6:].tolist() # 2016 ~ 2021

# 페이스북
plt.plot(x_values, facebook, marker='s', linestyle='--', color='blue', label='페이스북')
# 값 삽입: 값 + 0.25로 값 약간 위로 위치하게 조정
for i in range(len(x_values)):
    height = facebook[i]
    plt.text(x_values[i], height+0.25, '%.1f' %height, ha='center', va='bottom', size=15)

# 인스타그램
plt.plot(x_values, instagram, marker='s', linestyle='--', color='red', label='인스타그램')
# 값 삽입: 값 + 0.25로 값 약간 위로 위치하게 조정
for i in range(len(x_values)):
    height = instagram[i]
    plt.text(x_values[i], height+0.25, '%.1f' %height, ha='center', va='bottom', size=15)

# 카카오스토리
plt.plot(x_values, kakaostory, marker='s', linestyle='--', color='gold', label='카카오스토리')
# 값 삽입: 값 + 0.25로 값 약간 위로 위치하게 조정
for i in range(len(x_values)):
    height = kakaostory[i]
    plt.text(x_values[i], height+0.25, '%.1f' %height, ha='center', va='bottom', size=15)

# 네이버 밴드
plt.plot(x_values, band, marker='s', linestyle='--', color='green', label='네이버 밴드')
# 값 삽입: 값 + 0.25로 값 약간 위로 위치하게 조정
for i in range(len(x_values)):
    height = band[i]
    plt.text(x_values[i], height+0.25, '%.1f' %height, ha='center', va='bottom', size=15)

plt.title('2016년~2021년 서울시 SNS별 이용 현황', size = 30)
plt.legend(loc = (1, 0.6))
plt.show()
```

# 실습 2.
### 2021년 09월 서울시 전동킥보드 견인 현황 비교
https://data.seoul.go.kr/dataList/OA-21304/S/1/datasetView.do

#### 1. 라이브러리 import

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

#### 2. csv 파일 불러오기

```python
df = pd.read_csv('서울시 전동킥보드 견인 현황.csv', encoding='CP949')
df
```

#### 3. 2021년 9월 서울시 전동킥보드 견인 현황 데이터 추출

```python
# 2021년 9월 데이터 추출
df_09 = df.loc[df['신고일시'].str.contains('2021-09')]
df_09
```

```python
# '유형' 데이터 추출
tow = df_09[['유형']]

# 견인 유형별 데이터 그룹화
tow = tow.groupby('유형').size().reset_index(name='count').sort_values(by='count', ascending=False)
tow = tow.loc[tow['count'] > 10]

# 각 열을 리스트로 변환
tow_type = tow['유형'].tolist()
tow_count = tow['count'].tolist()
```

#### 4. 그래프 그리기

```python
plt.rcParams["font.family"] = "Malgun Gothic"

fig = plt.figure(figsize=(20, 20))

wedgeprops = {'width': 0.6, 'linewidth': 1, 'edgecolor': 'black'}
colors = ['gold', 'silver', 'cyan', 'white', 'pink', 'green', 'orange']
plt.pie(tow_count, labels=tow_type, autopct = '%.1f%%',colors=colors, shadow = True, wedgeprops = wedgeprops, pctdistance=0.7, textprops={'size':20})
plt.title('2021년 9월 서울시 전동 킥보드 견인 현황', size = 30)
plt.legend(loc = (1, 0.5), fontsize=15)
plt.show()
```

# 실습 3. 
### 2020년 국민건강보험 가입자 100명의 체중별 허레둘레 비교
https://www.data.go.kr/data/15007122/fileData.do

#### 1. 라이브러리 import

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

#### 2. csv 파일 불러오기

```python
df = pd.read_csv('국민건강보험공단_건강검진정보_20211229.CSV', encoding='CP949')
df
```

#### 3. 2020년 국민건강보험 가입자들의 키와 몸무게 비교

```python
# 2020년 국민건강보험 가입자 100명 무작위로 추출
df = df.sample(n=300)
df
```

```python
# 남성 데이터 추출
male = df.loc[df['성별코드'] == 1, ['체중(5Kg 단위)','허리둘레']]
male
```

```python
# 여성 데이터 추출
female = df.loc[df['성별코드'] == 2, ['체중(5Kg 단위)','허리둘레']]
female
```

```python
# 남성의 평균 체중 및 허리둘레
male.mean()
```

```python
# 여성의 평균 체중 및 허리둘레
female.mean()
```

#### 4. 그래프 그리기

```python
plt.rcParams["font.family"] = "Malgun Gothic"
plt.rc('font', size=15) # 기본 폰트 크기
plt.rc('axes', labelsize=25) # x, y축 label 폰트 크기
plt.rc('xtick', labelsize=20) # x축 눈금 폰트 크기
plt.rc('ytick', labelsize=20) # y축 눈금 폰트 크기

fig = plt.figure(figsize=(20,15))

plt.scatter(male['체중(5Kg 단위)'], male['허리둘레'], color='b', alpha=0.7, label='남성', marker='x', s = 150)
plt.scatter(female['체중(5Kg 단위)'], female['허리둘레'], color='r', alpha=0.6, label='여성', s = 150)

plt.xlabel('체중(kg)')
plt.ylabel('허리둘레(cm)')
plt.title("2020년 국민건강보험 가입자 300명의 체중별 허레둘레 비교", size = 30)

plt.legend()
plt.show()
```

# 실습 4.
### 2020년 건강보험 가입자 성별, 음주 여부별 체중 비교
https://www.data.go.kr/data/15007122/fileData.do

#### 1. 라이브러리 import

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

#### 2. csv 파일 불러오기

```python
df = pd.read_csv('국민건강보험공단_건강검진정보_20211229.CSV', encoding='CP949')
df
```

#### 3. 2020년 건강보험 가입자의 성별 및 음주여부 데이터 변환

```python
# 성별 데이터 (1, 2) -> (남성, 여성) 변환
df['성별코드'] = df['성별코드'].map({1:'남성',
                                     2:'여성'})
df.head(10)
```

```python
# 음주 여부 데이터 (0, 1) -> ('안 마심', '마심')
df['음주여부'] = df['음주여부'].map({0:'안 마심',
                                     1:'마심'})
df.head(10)
```

#### 4. 그래프 그리기

```python
plt.rcParams["font.family"] = "Malgun Gothic"
plt.rc('font', size=15) # 기본 폰트 크기
plt.rc('axes', labelsize=25) # x, y축 label 폰트 크기
plt.rc('xtick', labelsize=20) # x축 눈금 폰트 크기
plt.rc('ytick', labelsize=20) # y축 눈금 폰트 크기

fig = plt.figure(figsize=(20,15))

sns.catplot(x='음주여부', y ='체중(5Kg 단위)', hue='성별코드', kind='boxen', data=df, height=10, palette=sns.color_palette(['blue', 'red']))
plt.title("2020년 건강보험 가입자의 성별, 음주 여부별 체중 비교", size=25)
```

# 실습 5.
### 서울특별시 광진구 연령별 인구분포
https://www.data.go.kr/data/15052328/fileData.do

#### 1. 라이브러리 import

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

#### 2. csv 파일 불러오기

```python
df = pd.read_csv('서울특별시 광진구_연령별인구현황_20220131.csv', encoding='CP949')
df
```

#### 3. 서울특별시 광진구 연령별 인구 데이터 추출

```python
population = df['광진구'][0:101].tolist()
for i in range(len(population)):
    population[i] = int(population[i].replace(",","")) # 문자열 -> 정수 변환 및 천 단위 콤마 제거
population
```

#### 4. 그래프 그리기

```python
plt.rcParams["font.family"] = "Malgun Gothic"
plt.rc('font', size=15) # 기본 폰트 크기
plt.rc('axes', labelsize=25) # x, y축 label 폰트 크기
plt.rc('xtick', labelsize=20) # x축 눈금 폰트 크기
plt.rc('ytick', labelsize=20) # y축 눈금 폰트 크기

fig = plt.figure(figsize=(20,15))

age = list(np.arange(101))

plt.style.use('ggplot') # 격자무니 스타일
plt.plot(age, population, color='b')
plt.xlabel('나이')
plt.ylabel('인구 수')
plt.title("서울특별시 광진구 연령별 인구 분포", size = 30)

dic = { y:x for x, y in zip(age, population) }
plt.text(dic[min(dic)],min(dic),
         str(dic[min(dic)]) + '살 - Min Point: ' + str(min(dic)),
         color='r',
         horizontalalignment='center',
         verticalalignment='top',
         fontsize=20
        )
plt.text(dic[max(dic)],max(dic),
         str(dic[max(dic)]) + '살 - Max Point: ' + str(max(dic)),
         color='r',
         horizontalalignment='center',
         verticalalignment='bottom',
         fontsize=20
        )

plt.show()
```

# 실습 6.
### 2021년 광진구 연령별 1인가구 통계
https://data.seoul.go.kr/dataList/10995/S/2/datasetView.do?stcSrl=10995

#### 1. 라이브러리 import

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

#### 2. csv 파일 불러오기

```python
df = pd.read_csv('1인가구(연령별)_20221004052012.csv')
df
```

#### 3. 광진구 연령별 1인가구 데이터 추출

```python
# 연령 label 추출
age = df.iloc[1, 6::3].tolist()
age
```

```python
# 남성 1인가구수 데이터 추출
gwangjin_gu_male = df.loc[df['자치구별(2)']=='광진구'].iloc[:, 6::3].values.flatten().tolist()
gwangjin_gu_male = list(map(int, gwangjin_gu_male))
gwangjin_gu_male
```

```python
# 여성 1인가구수 데이터 추출
gwangjin_gu_female = df.loc[df['자치구별(2)']=='광진구'].iloc[:, 7::3].values.flatten().tolist()
gwangjin_gu_female = list(map(int, gwangjin_gu_female))
gwangjin_gu_female
```

#### 4. 그래프 그리기

```python
plt.rcParams["font.family"] = "Malgun Gothic"
plt.rc('font', size=15)


fig = plt.figure(figsize=(20,15))

plt.subplot(2, 1, 1)
plt.barh(age, gwangjin_gu_male, color='b')
plt.barh(age, gwangjin_gu_female, left=gwangjin_gu_male, color='lightgray')
plt.title("2021년 서울시 광진구 연령별 1인가구 수", size = 30)
plt.xlabel("1인가구 수")
plt.legend(['남성', '전체'])

for i, v in enumerate(age):
    str_val = "%d" % gwangjin_gu_male[i]
    plt.text(gwangjin_gu_male[i], v, str_val, ha='left', va='center', fontsize=15, color='#3300ff', fontweight='bold')
    
plt.subplot(2, 1, 2)
plt.barh(age, gwangjin_gu_female, color='r', label='여성')
plt.barh(age, gwangjin_gu_male, left=gwangjin_gu_female, color='lightgray')
plt.xlabel("1인가구 수")
plt.legend(['여성', '전체'])

for i, v in enumerate(age):
    str_val = "%d" % gwangjin_gu_female[i]
    plt.text(gwangjin_gu_female[i], v, str_val, ha='left', va='center', fontsize=15, color='#ff3300', fontweight='bold')



plt.show()
```
